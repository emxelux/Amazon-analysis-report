```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Amazon Sale Report - Copy.csv")
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop(['index', 'fulfilled-by', 'Unnamed: 22'], axis = 1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['ship-country'].unique()
```

```python
df['ship-country'] = df['ship-country'].fillna(df['ship-country'].mode([0]))
```

```python
df.isnull().sum()
```

```python
df.Status.unique()
```

```python
df['Status'] = df['Status'].replace({
    'Shipped - Delivered to Buyer': 'Delivered',
    'Shipped - Returned to Seller': 'Returned',
    'Shipped - Rejected by Buyer': 'Rejected',
    'Shipped - Lost in Transit': 'Lost in Transit',
    'Shipped - Out for Delivery': 'In Transit',
    'Shipped - Returning to Seller': 'Returned', # Assuming it's on its way back, but not fully 'Returned' yet
    'Shipped - Picked Up': 'In Transit',
    'Shipped - Damaged': 'Damaged',
    'Shipping': 'In Transit',
    'Pending - Waiting for Pick Up': 'Pending',
    'Shipped': 'In Transit' # Generic 'Shipped' often means in transit
})
```

```python
print(df['Status'].unique())
```

```python
df.isnull().sum()
```

```python
print(df['Courier Status'].unique())
```

```python
# Filling null values in 'Courier Status' based on 'Status' column
df.loc[df['Courier Status'].isnull(), 'Courier Status'] = df.loc[df['Courier Status'].isnull(), 'Status']

# Standardized some 'Courier Status' values to match 'Status' categories
df['Courier Status'] = df['Courier Status'].replace({
    'Shipped': 'In Transit', # Map generic 'Shipped' to 'In Transit'
    'Cancelled': 'Cancelled',
    'Pending': 'Pending',
    'Returned': 'Returned',
    'Rejected': 'Rejected',
    'Lost in Transit': 'Lost in Transit',
    'Damaged': 'Damaged',
    'Delivered': 'Delivered'
})

print(df['Courier Status'].unique())
print(df['Courier Status'].isnull().sum())
```

```python
df.isnull().sum()
```

```python
df.currency.unique()
```

```python
df.currency.fillna(df.currency.mode()[0], inplace=True)
```

```python
df.head()
```

```python
df[df.Amount <=0]
```

```python
df.isnull().sum()
```

```python
cancelled_zero_amount_orders = df[(df['Amount'] <= 0) & (df['Status'] == 'Cancelled')]
print(f"Number of cancelled orders with 0 or less amount: {len(cancelled_zero_amount_orders)}")
display(cancelled_zero_amount_orders[['Status', 'Courier Status', 'Amount']].head())
```

```python
df.head()
```

```python
df.drop(['currency', 'Courier Status', 'ship-country'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Total Amount'] = df['Amount'] * df['Qty']
```

```python
df.head()
```

```python
print(f"Missing values in 'Amount' column: {df['Amount'].isnull().sum()}")
```

```python
df['Amount'] = df.groupby('SKU')['Amount'].transform(lambda x: x.fillna(x.median()))
print(f"Missing values in 'Amount' column after SKU-based imputation: {df['Amount'].isnull().sum()}")
```

```python
df['Amount'] = df.groupby('Category')['Amount'].transform(lambda x: x.fillna(x.median()))
print(f"Missing values in 'Amount' column after Category-based imputation: {df['Amount'].isnull().sum()}")
```

```python
print(f"Missing values in 'Amount' column after all imputations: {df['Amount'].isnull().sum()}")
df.head()
```

```python
df.isnull().sum()
```

```python
list(df['ship-city'].unique())
```

```python
df['ship-city']=df['ship-city'].fillna("Unknown Cities")
```

```python
df.isnull().sum()
```

```python
df[df['ship-state'].isnull()].head(3)
```

```python
df['ship-state'] = df['ship-state'].fillna("Unknown")
df['ship-postal-code'] = df['ship-postal-code'].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['promotion-ids'] = df['promotion-ids'].fillna("No Promotions")
df['Total Amount'] = df['Amount'] * df['Qty']
```
